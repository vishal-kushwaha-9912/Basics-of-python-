#### 1. Square Patterns

#### ⭐ Solid Square

````python
for x in range(1,6):
    for y in range(1,6):
        print("*",end=" ")
    print()
````

#### ⭐ Number Square

````python
for x in range(1,6):
    for y in range(1,6):
        print(y,end=" ")
    print()
````

#### ⭐ Alphabet Square

````python
for x in range(1,6):
    for y in range(1,6):
        print(chr(64+y),end=" ")
    print()
````

### 3.Half Pyramid

#### Left Half Star Pyramid


#### ⭐ Left Half Star Pyramid

````python
for x in range(1, 7):
    for y in range(1, x + 1):
        print("*", end=" ")
    print()
````

#### ⭐ Inverted Half Star Pyramid


````python
for x in range(7, 0, -1):
    for y in range(1, x + 1):
        print("*", end=" ")
    print()
````

#### ⭐ Number Half Pyramid


````python
for x in range(1, 6):
    for y in range(1, x + 1):
        print(y, end=" ")
    print()
````

#### ⭐ Repeated Number Pyramid


````python
for x in range(1, 6):
    for y in range(1, x + 1):
        print(x, end=" ")
    print()
````

#### ⭐ Alphabet Half Pyramid


````python
for x in range(0, 6):
    for y in range(0, x):
        print(chr(y + 65), end=" ")
    print()
````

#### ⭐ Repeated Alphabet Pyramid

````python
for x in range(1, 6):
    for y in range(0, x):
        print(chr(x + 64), end=" ")
    print() 
````

#### ⭐ Continuous Numbers

````python
num = 1
for x in range(1, 6):
    for y in range(1, x+1):
        print(num, end=" ")
        num +=1
    print()
````

### Full Pyramid

#### ⭐ Star Pyramid

````python
for x in range(1, 6):
    space = 5 - x
    for y in range(0, space):
        print(" ", end="")
    for y in range(0, x):
        print("*", end=" ")
    print()
````

#### ⭐ Number Pyramid

````python
for x in range(1, 7):
    space = 6 - x
    for y in range(0, space):
        print(" ", end="")
    for y in range(1, x):
        print(y, end=" ")
    print()
````

````python
for x in range(1, 6):
    space = 5 - x
    for y in range(0, space):
        print(" ", end="")
    for y in range(0, x):
        print(x, end=" ")
    print()
````

### Diamond Patterns

##### ⭐ Solid Diamond

````python


for x in range(1, 5):
    space = 4 - x
    for y in range(0, space):
        print("", end=" ")
    for y in range(1, x + 1):
        print("*", end=" ")
    print()

    # inverse

for x in range(3, 0, -1):
    space = 4 - x

    for y in range(space):
        print("", end=" ")

    for y in range(x):
        print("*", end=" ")

    print()
````

#### ⭐ Number Diamond

````python


for x in range(1, 6):
    space = 5- x
    for y in range(0, space):
        print("", end=" ")
    for y in range(1, x + 1):
        print(y, end=" ")
    print()

    # inverse

for x in range(4,0,-1):
    space = 4-x 

    for y in range(space):
        print("", end=" ")

    for y in range(1,x+1):
        print(y,end=" ")

    print()
````

### 7. Hollow Patterns

#### ⭐Hollow square

````python

for x in range (1,5):
     for y in range(1,5):
       if x == 1 or x==4 or y==1 or y==4:
        print("*", end=" ")
       else:
         print(" ",end=" ")
        
     print()



# for x in range (1,5):
#      for y in range(1,5):
#         print("*", end=" ")
#      print()

#OUTPUT::  
# * * * * 
# * * * * 
# * * * * 
# * * * * 
# now we remove the inner star of the square like this 
# * * * * 
# *     * 
# *     * 
# * * * *     the result of the question is <<<<<<<<
# think how to remove the inner stars from the square
# if we remove
#        ▽
# x==1 row 
# x==4 row
# y==1 col
# y==4 col 
````
