### Half Pyramid

#### Left Half Star Pyramid


```python
for x in range(1, 7):
    for y in range(1, x + 1):
        print("*", end=" ")
    print()
```

#### Inverted Half Star Pyramid


```python
for x in range(7, 0, -1):
    for y in range(1, x + 1):
        print("*", end=" ")
    print()
```

#### ⭐ Number Half Pyramid


```python
for x in range(1, 6):
    for y in range(1, x + 1):
        print(y, end=" ")
    print()
```

#### ⭐ Repeated Number Pyramid


```python
for x in range(1, 6):
    for y in range(1, x + 1):
        print(x, end=" ")
    print()
```

#### ⭐ Alphabet Half Pyramid


```python
for x in range(0, 6):
    for y in range(0, x):
        print(chr(y + 65), end=" ")
    print()
```

#### ⭐ Repeated Alphabet Pyramid

```python
for x in range(1, 6):
    for y in range(0, x):
        print(chr(x + 64), end=" ")
    print() 
```

#### ⭐ Continuous Numbers

```python
for x in range(1, 6):
    for y in range(1, x+1):
        print(x, end=" ")
    print()
```

### Full Pyramid

#### ⭐ Star Pyramid

```python
for x in range(1, 6):
    space = 5 - x
    for y in range(0, space):
        print(" ", end="")
    for y in range(0, x):
        print("*", end=" ")
    print()
```

#### ⭐ Number Pyramid

```python
for x in range(1, 7):
    space = 6 - x
    for y in range(0, space):
        print(" ", end="")
    for y in range(1, x):
        print(y, end=" ")
    print()
```
