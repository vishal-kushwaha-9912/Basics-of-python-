```python
# BREAK STATEMENT :A break statement is used to immediately stop a loop and come out of it.



# Real-Life Example

# Imagine you're searching for your friend in a classroom.

# Student 1 ❌
# Student 2 ❌
# Student 3 ❌
# Friend Found ✅
# Stop Searching

# As soon as you find your friend, you stop searching.

# This "stop searching" is like break.



for i in range(1,11):
    if i==5:
        break
    print(i)
```
