## 7. min/max

#### The min() function returns the smallest value.
#### The max() function returns the largest value.

```python
min([2,4,21,5,8])

#max([21,24,31,4,8])  output is 31
```

```python
print(min("apple", "banana", "cherry"))
print(max("apple", "banana", "cherry"))

```

```python
print(min("python"))
print(max("python"))
```

```python
fruits = ["mango", "apple", "banana"]

print(min(fruits))
print(max(fruits))
```

## 8. round

#### The round() function is used to round a number to the nearest integer or a specified number of decimal places.

```python
print(round(4.6)) #after 0.5 its count in next number 
print(round(4.2))
print(round(3.14159, 2)) # here 2 is define number of digits after decimal you want.print(round(1234, -1))
print(round(1234, -2))
```

## 9. divmod 

#### The divmod() function returns both quotient and remainder when one number is divided by another.

```python
# Instead of writing:
# quotient = 17 // 5 => 3
# remainder = 17 % 5 => 2

# you can use:
# divmod(quotient, remainder)

# divmod(17, 5)

q, r = divmod(17, 5)

print("Quotient =", q)
print("Remainder =", r)
```

```python
friends = 5
chocolates = 23

q, r = divmod(chocolates, friends)

print("Each friend gets:", q)
print("Leftover chocolates:", r)
```

## 11. bin/oct /hext

#### These functions are used to convert a decimal (base 10) number into:
<!-- 
bin() → Binary (base 2)
oct() → Octal (base 8)
hex() → Hexadecimal (base 16) -->

```python
num = 25

print("Binary:", bin(num))
print("Octal:", oct(num))
print("Hexadecimal:", hex(num))
```
