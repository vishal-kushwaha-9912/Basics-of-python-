#### A file containing a set of funcction you want to include in your application .

* Math
* Random

##### **Note:** Before using functions from a module such as math, time, os, or random, you must first import the module. After importing, you can access its functions using the dot (.) operator.

| ------------------ | ----------------------- | ------------------- | ------------------- |
| `math.sqrt()`      | Square root             | `math.sqrt(25)`     | `5.0`               |

| `math.pow()`       | Power                   | `math.pow(2, 3)`    | `8.0`               |
| `math.gcd()`       | Greatest Common Divisor | `math.gcd(12, 18)`  | `6`                 |
| `math.fabs()`      | Absolute value          | `math.fabs(-7)`     | `7.0`               |
| `math.pi`          | Value of π              | `math.pi`           | `3.141592653589793` |
| `math.e`           | Euler's number          | `math.e`            | `2.718281828459045` |
| `math.sin()`       | Sine value              | `math.sin(0)`       | `0.0`               |

* TIME 


| Function              | Purpose              | Example                                  | Sample Output                                               |
| --------------------- | -------------------- | ---------------------------------------- | ----------------------------------------------------------- |
| `time.time()`         | Current timestamp    | `time.time()`                            | `1750500000.123`                                            |
| `time.sleep()`        | Pause program        | `time.sleep(2)`                          | Waits 2 seconds                                             |
| `time.ctime()`        | Current date & time  | `time.ctime()`                           | `Sun Jun 21 15:30:00 2026`                                  |
| `time.localtime()`    | Local time details   | `time.localtime()`                       | `time.struct_time(tm_year=2026, tm_mon=6, ...)`             |
| `time.gmtime()`       | UTC/GMT time details | `time.gmtime()`                          | `time.struct_time(tm_year=2026, tm_mon=6, ...)`             |
| `time.strftime()`     | Format date/time     | `time.strftime("%d-%m-%Y")`              | `21-06-2026`                                                |
| `time.strptime()`     | String → Time object | `time.strptime("21-06-2026","%d-%m-%Y")` | `time.struct_time(tm_year=2026, tm_mon=6, tm_mday=21, ...)` |
| `time.perf_counter()` | High-precision timer | `time.perf_counter()`                    | `1234.56789`                                                |
| `time.process_time()` | CPU processing time  | `time.process_time()`                    | `0.125`                                                     |


```python
import time
time.time()
# find the total second
```

```python
time.ctime() # current time 
```

* OS 
#### The os module is a built-in module used to interact with the Operating System.




| Function           | Purpose                       | Example                          | Sample Output                    |
| ------------------ | ----------------------------- | -------------------------------- | -------------------------------- |
| `os.getcwd()`      | Get current working directory | `os.getcwd()`                    | `C:\Users\Vishal\Desktop\python` |
| `os.chdir()`       | Change directory              | `os.chdir("D:/Python")`          | Directory changed                |
| `os.listdir()`     | List files/folders            | `os.listdir()`                   | `['test.py', 'data.txt']`        |
| `os.mkdir()`       | Create a folder               | `os.mkdir("Demo")`               | Folder created                   |
| `os.rmdir()`       | Remove empty folder           | `os.rmdir("Demo")`               | Folder removed                   |
| `os.remove()`      | Delete a file                 | `os.remove("test.txt")`          | File deleted                     |
| `os.rename()`      | Rename file/folder            | `os.rename("old.txt","new.txt")` | Renamed                          |
| `os.path.exists()` | Check if path exists          | `os.path.exists("test.txt")`     | `True`                           |
| `os.system()`      | Run system command            | `os.system("dir")`               | Displays files                   |
| `os.name`          | Operating system name         | `os.name`                        | `nt`                             |


```python
import os
```
