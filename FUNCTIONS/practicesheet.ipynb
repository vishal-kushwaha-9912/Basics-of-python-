### 2. Built-in Function Question

Create a list:

numbers = [12, 5, 8, 20, 3]

Use built-in functions to print:

Length of list (len())
Largest number (max())
Smallest number (min())
Sorted list (sorted())

```python

number = [12 , 5, 8,20 ,3]

length  = (len(number))
largest = (max(number))
smallest = (min(number))
sorted_list = (sorted(number))

print("length" , length)
print("largest" , largest)
print("smallest" , smallest)
print("sortest list ",sorted_list)
```
