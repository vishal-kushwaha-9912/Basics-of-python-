#### Question 1 – Built-in Functions

Create a list:

numbers = [25, 12, 8, 40, 17]

Print:

Length of the list
Sum of all numbers
Largest number
Smallest number

```python
import math
number = [25 , 12 , 8 , 40 , 17 ]

length =len(number)
total = sum(number)
largest = max(number)
smallest = min(number)


print ("length is :" ,length)
print ("total is :" ,total)
print ("the largest number is :", largest)
print ("the smallest number is : ",  smallest)
```

#### Question 2 – Random Module

Generate a random number between 1 and 50.

Print:

The random number
Its square using **
Whether it is Even or Odd

```python
import random
number=random.randint(1,50)


square = number**2
print("The square : ", square)
print ("The random number : ", number)
```
