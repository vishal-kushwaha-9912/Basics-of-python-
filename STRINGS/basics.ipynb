### Strings

#### A String (str) is a data type in Python used to store text. A string is a sequence of characters enclosed inside single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """). Python identifies strings with the str data type.

#### What is a Character?

#### A character can be:

* A letter → A, b, z
* A number → 1, 5, 9
* A symbol → @, #, $
* A space → " "

```python
name = 'Vishal'  # Single Quotes
city = "Delhi"    #Double Quotes
message ='''Hello           
My name is Vishal
I am learning Python'''  # Triple Quotes (Multi-line String)

print(name,city,message , sep ="\n" )
```

####  * Python allows single quotes (' '), double quotes (" "), and triple quotes (''' ''' or """ """)
##### to make writing strings easier and more flexible.

##### Example:
print("Hello Vishal, it's me Anmol. Do you remember me?")

##### In the example above, the string contains an apostrophe (') in "it's".
#### Since the string is enclosed in double quotes (" "), Python knows that
##### the apostrophe is part of the text and not the end of the string.

##### If we used single quotes, Python would get confused:
##### print('Hello Vishal, it's me Anmol. Do you remember me?')  ❌ Error

##### Triple quotes are useful for writing multi-line strings and documentation.

##### Having three types of quotes makes code easier to write, read, and understand.

## String Methods

| Method         | What it does                  |
| -------------- | ----------------------------- |
| `upper()`      | Converts to uppercase         |
| `lower()`      | Converts to lowercase         |
| `strip()`      | Removes spaces from both ends |
| `replace()`    | Replaces text                 |
| `find()`       | Finds the first index of text |
| `count()`      | Counts occurrences            |
| `startswith()` | Checks the beginning          |
| `endswith()`   | Checks the ending             |
| `split()`      | Converts a string into a list |
| `join()`       | Converts a list into a string |


| Method         | Easy Meaning                               |
| -------------- | ------------------------------------------ |
| `upper()`      | **UP** → Make all letters CAPITAL          |
| `lower()`      | **LOW** → Make all letters small           |
| `strip()`      | **STRIP** → Remove extra spaces            |
| `replace()`    | Replace one word with another              |
| `find()`       | Find where a word/letter is                |
| `count()`      | Count how many times it appears            |
| `startswith()` | Starts with...? (`True`/`False`)           |
| `endswith()`   | Ends with...? (`True`/`False`)             |
| `split()`      | Split one string into many pieces (a list) |
| `join()`       | Join many pieces (a list) into one string  |


```python
text = "  I Love Python  "

print(text.upper())              # I LOVE PYTHON
print(text.lower())              # i love python
print(text.strip())              # I Love Python
print(text.replace("Python", "Java"))  # I Love Java
print(text.find("Love"))         # 4
print(text.count("o"))           # 2
print(text.startswith("  I"))    # True
print(text.endswith("  "))       # True

languages = "Python,Java,C++"
print(languages.split(","))      # ['Python', 'Java', 'C++']

words = ["I", "Love", "Python"]
print(" ".join(words))           # I Love Python
```
