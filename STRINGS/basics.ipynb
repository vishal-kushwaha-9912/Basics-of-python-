### Strings

#### A String (str) is a data type in Python used to store text. A string is a sequence of characters enclosed inside single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """). Python identifies strings with the str data type.

#### What is a Character?

#### A character can be:

* A letter → A, b, z
* A number → 1, 5, 9
* A symbol → @, #, $
* A space → " "

```python
name = 'Vishal'  # Single Quotes
city = "Delhi"    #Double Quotes
message ='''Hello           
My name is Vishal
I am learning Python'''  # Triple Quotes (Multi-line String)

print(name,city,message , sep ="\n" )
```

####  * Python allows single quotes (' '), double quotes (" "), and triple quotes (''' ''' or """ """)
##### to make writing strings easier and more flexible.

##### Example:
print("Hello Vishal, it's me Anmol. Do you remember me?")

##### In the example above, the string contains an apostrophe (') in "it's".
#### Since the string is enclosed in double quotes (" "), Python knows that
##### the apostrophe is part of the text and not the end of the string.

##### If we used single quotes, Python would get confused:
##### print('Hello Vishal, it's me Anmol. Do you remember me?')  ❌ Error

##### Triple quotes are useful for writing multi-line strings and documentation.

##### Having three types of quotes makes code easier to write, read, and understand.
