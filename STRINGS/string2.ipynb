#### Accessing Substrings from a String in Python

#### A substring is a part (portion) of a string. In Python, we use slicing to access substrings.

#### Concept of Indexing in Python

#### Indexing is the process of accessing individual characters of a string using their position number (index).

##### word = "Python"

| Character | P | y | t | h | o | n |
| --------- | - | - | - | - | - | - |
| Index     | 0 | 1 | 2 | 3 | 4 | 5 |


##### Why Indexing?
##### Indexing allows you to:

* Access a specific character
* Extract parts of a string
* Manipulate text efficiently
