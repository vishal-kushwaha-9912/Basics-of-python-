This means that once a string is created, its characters cannot be changed.

# DELETION 
### Can we delete characters from a string?

##### No. Since strings are immutable, you cannot delete individual characters from a string after it is created.


❌ Incorrect Example
name = "Python"

del name[0]

Output:

TypeError: 'str' object doesn't support item deletion

Why?

* A string is immutable.
* You cannot change, replace, or delete individual characters.

✅ Correct Way

##### If you want to remove a character, create a new string.

name = "Python"

new_name = name[1:]

print(new_name)

Output:

ython

#### Here, the original string "Python" is not changed. Python creates a new string "ython".
