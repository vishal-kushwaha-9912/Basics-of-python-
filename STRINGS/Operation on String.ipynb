✅ Small (lowercase) letters are greater than capital (uppercase) letters in ASCII.
So, 
   Small letters are Big

```python
print('A' < 'a')   # True
print('Z' < 'a')   # True
print('a' > 'A')   # True
print(ord('A'))    # 65
print(ord('a'))    # 97
```

5️⃣ Loop Operations

```python
v = "vishal kushwaha"
for x in v:
    print(x)
```

```python
v = "vishal kushwaha"
for x in v[0:9]: # you can also define how many alphabets print
    print(x)
```
