5. format

## 5. `format()`

### Definition
It is used to insert values into a string by replacing placeholders `{}`.

### Syntax
```python
string.format(value1, value2, ...)
```


---

### Note
- `{}` are called **placeholders**.
- `format()` replaces the placeholders with the given values.
- Values are inserted in the same order unless you use indexes or names.

````python
"My name is {} and I am {} years old.".format("vishal",18)
````

````python
"My name is {1} and I am {0} years old.".format("vishal",18)
````

````python
print("My name is {name} and I am {age} years old.".format(name="vishal",age= 18,weight = 67))
````

#### How `format()` Works

When we use `format()`, we pass the values inside the `format()` method.

```python
print("My name is {0} and I am {1} years old.".format("Vishal", 18))
```

The values inside `format()` are stored in order:

| Value | Position (Index) |
|--------|------------------|
| `"Vishal"` | `0` |
| `18` | `1` |

The numbers inside `{}` are called **placeholders**.

- `{0}` displays the value at index `0`.
- `{1}` displays the value at index `1`.

**Output**

```text
My name is Vishal and I am 18 years old.
```

---

### What happens if we change the positions?

If we change the placeholder positions, the values are printed according to the new positions.

```python
print("My name is {1} and I am {0} years old.".format("Vishal", 18))
```

**Output**

```text
My name is 18 and I am Vishal years old.
```

Here,
- `{1}` prints the value at index `1` → `18`
- `{0}` prints the value at index `0` → `"Vishal"`

So, the values are **swapped**.

---

### Another Example

```python
print("{2} {0} {1}".format("Python", "Programming", "Learn"))
```

**Output**

```text
Learn Python Programming
```

Explanation:
- `{2}` → `"Learn"`
- `{0}` → `"Python"`
- `{1}` → `"Programming"`

---

### Note
- The first value passed to `format()` has index `0`.
- The second value has index `1`.
- You can use the same index multiple times.

```python
print("{0} {0} {1}".format("Hello", "World"))
```

**Output**

```text
Hello Hello World
```

- If you don't write the indexes, Python fills the placeholders from left to right.

```python
print("{} {}".format("Hello", "World"))
```

**Output**

```text
Hello World
```

````python
name = "Vishal"
age = 20

print("My name is {} and I am {} years old.".format(name, age))
````

````python
fruit = "Apple"
price = 120

print("The price of {} is ₹{}.".format(fruit, price))
````

## 10. `isidentifier()`

### Definition
It is used to check whether a string is a **valid Python identifier (variable name)**.

### Syntax
```python
string.isidentifier()
```

### Example

```python
print("name".isidentifier())
print("student_name".isidentifier())
print("2name".isidentifier())
print("student-name".isidentifier())
```

### Output

```
True
True
False
False
```

### Rules for a Valid Identifier

- Must start with a letter (`A-Z` or `a-z`) or an underscore (`_`).
- Cannot start with a number.
- Can contain letters, numbers, and underscores.
- Cannot contain spaces or special characters like `@`, `#`, `-`, `%`, etc.

### Examples

| String | Result |
|--------|--------|
| `"name"` | ✅ True |
| `"_age"` | ✅ True |
| `"student1"` | ✅ True |
| `"1student"` | ❌ False |
| `"student-name"` | ❌ False |
| `"student name"` | ❌ False |

## 7. `split()`

### Definition
It is used to split a string into a list using a specified separator.

### Syntax
```python
string.split(separator)
```

If no separator is given, Python splits the string at **spaces** by default.

---

### Example 1 (Default Split)

```python
text = "Python is easy to learn"

print(text.split())
```

### Output

```python
['Python', 'is', 'easy', 'to', 'learn']
```

---

### Example 2 (Split by Space)

```python
text = "Python is easy to learn"

print(text.split(" "))
```

### Output

```python
['Python', 'is', 'easy', 'to', 'learn']
```

---

### Example 3 (Split by Comma)

```python
text = "Apple,Banana,Mango,Orange"

print(text.split(","))
```

### Output

```python
['Apple', 'Banana', 'Mango', 'Orange']
```

---

### Example 4 (Split by Hyphen)

```python
text = "2026-07-09"

print(text.split("-"))
```

### Output

```python
['2026', '07', '09']
```

---

### Example 5 (Split by a Word)

```python
text = "Python-is-fun"

print(text.split("-"))
```

### Output

```python
['Python', 'is', 'fun']
```

---

### Note
- `split()` always returns a **list**.
- By default, it splits at spaces.
- You can use any separator such as `" "`, `","`, `"-"`, `":"`, etc.
- If the separator is not found, the original string is returned as a single item in a list.

### Example

```python
text = "Python"

print(text.split(","))
```

### Output

```python
['Python']
```

## 8. `join()`

### Definition
It is used to join the elements of a list (or any iterable) into a single string using a specified separator.

### Syntax
```python
separator.join(iterable)
```

- **separator** → The character or string placed between the elements.
- **iterable** → A list, tuple, or other sequence of strings.

---

### Example 1 (Join with Space)

```python
words = ["Python", "is", "easy"]

print(" ".join(words))
```

### Output

```
Python is easy
```

---

### Example 2 (Join with Hyphen)

```python
words = ["Python", "is", "easy"]

print("-".join(words))
```

### Output

```
Python-is-easy
```

---

### Example 3 (Join with Comma)

```python
fruits = ["Apple", "Banana", "Mango"]

print(", ".join(fruits))
```

### Output

```
Apple, Banana, Mango
```

---

### Example 4 (Join Without Separator)

```python
letters = ["P", "Y", "T", "H", "O", "N"]

print("".join(letters))
```

### Output

```
PYTHON
```

---

### Note
- `join()` returns a **string**.
- All elements in the list must be **strings**.
- If the list contains numbers, convert them to strings first.

### Example

```python
numbers = ["1", "2", "3"]

print("-".join(numbers))
```

### Output

```
1-2-3
```

---

### Common Error

```python
numbers = [1, 2, 3]

print("-".join(numbers))
```

### Output

```
TypeError: sequence item 0: expected str instance, int found
```

### Correct Way

```python
numbers = [1, 2, 3]

print("-".join(map(str, numbers)))
```

### Output

```
1-2-3
```

## 9. `replace()`

### Definition
It is used to replace a character, word, or substring in a string with another character, word, or substring.

### Syntax
```python
string.replace(old, new)
```

- **old** → The value to be replaced.
- **new** → The new value that replaces the old value.

---

### Example 1 (Replace a Word)

```python
text = "I love Python"

print(text.replace("Python", "Java"))
```

### Output

```
I love Java
```

---

### Example 2 (Replace a Character)

```python
text = "banana"

print(text.replace("a", "@"))
```

### Output

```
b@n@n@
```

---

### Example 3 (Replace a Substring)

```python
text = "Python is easy"

print(text.replace("easy", "powerful"))
```

### Output

```
Python is powerful
```

---

### Example 4 (Word Not Found)

```python
text = "Python"

print(text.replace("Java", "C++"))
```

### Output

```
Python
```

---

### Note
- `replace()` returns a **new string**.
- The original string does **not** change.
- If the old value is not found, the original string is returned.

## 10. `strip()`

### Definition
It is used to remove whitespace (spaces, tabs, and newlines) from the beginning and end of a string.

### Syntax
```python
string.strip()
```

---

### Example 1 (Remove Spaces)

```python
text = "   Python Programming   "

print(text.strip())
```

### Output

```
Python Programming
```

---

### Example 2 (Remove Specific Characters)

```python
text = "###Python###"

print(text.strip("#"))
```

### Output

```
Python
```

---

### Example 3 (Remove Multiple Characters)

```python
text = "@#$Python$#@"

print(text.strip("@#$"))
```

### Output

```
Python
```

---

### Example 4 (Characters Not at the Ends)

```python
text = "Py#thon"

print(text.strip("#"))
```

### Output

```
Py#thon
```

### Explanation
`strip()` removes characters **only from the beginning and the end**, not from the middle of the string.

---

### Note
- `strip()` returns a **new string**.
- The original string does **not** change.
- By default, it removes whitespace.
- You can also specify which characters to remove.
