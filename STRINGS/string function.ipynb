## 2. `count()`

### Definition
It is used to count how many times a specific character, word, or substring appears in a string.

### Syntax
```python
string.count(value)
```
### Note
- It returns an integer.
- If the value is not found, it returns `0`.

````python
text = "programming"
print(text.count("m"))
print(text.count("p"))
print(text.count("pro"))
print(text.count("ing"))
````

## 3. `find()`

### Definition
It is used to find the index (position) of the first occurrence of a character or substring in a string.

### Syntax
```python
string.find(value)
```
### If the value is not found
```python
text = "programming"

print(text.find("z"))
```

### Output
```
-1
```

### Note
- Returns the index of the first occurrence.
- Indexing starts from `0`.
- Returns `-1` if the value is not found.

````python
"programming".find("g")
````

````python
text = "programming"

print(text.find("z"))
print(text.find("m"))
````

## 4. `startswith()`

### `startswith()`

#### Definition
It is used to check whether a string starts with a specific character or substring.

### Syntax
```python
string.startswith(value)
```

### Example
```python
text = "Python Programming"

print(text.startswith("Python"))
print(text.startswith("Java"))
```

### Output
```
True
False
```

---

---

### Note
-  methods return a Boolean value (`True` or `False`).
- `startswith()` checks the beginning of the string.
- `endswith()` checks the end of the string.

````python
p="python java c++"
p.startswith("python")
# or
print(p.startswith("python"))
````

````python
text = "programming is used to change the world"

print(text.startswith("programming"))  # True
print(text.startswith("Programming"))  # False (case-sensitive)  P IS Capital letter
print(text.startswith("change"))       # False
````

````python
"programming is used to change the world".startswith("pro")
````

````python
print("programming is used to change the world".startswith("is used"))
````

````python
print("programming is used to change the world".startswith("change"))
````


### 4.`endswith()`

#### Definition
It is used to check whether a string ends with a specific character or substring.

### Syntax
```python
string.endswith(value)
```

### Example
```python
text = "Python Programming"

print(text.endswith("Programming"))
print(text.endswith("Python"))
```

### Output
```
True
False
```
---

### Note
-  methods return a Boolean value (`True` or `False`).
- `endswith()` checks the end of the string.

````python
text = "programming is used to change the world"

print(text.endswith("programming"))  # False
print(text.endswith("change"))  # False (case-sensitive)
print(text.endswith("world"))       # True
print(text.endswith("world"))       # True
````

````python
text = "Python"

print(text.endswith("Python"))  # True
print(text.endswith("python"))  # False
"""Here, the second result is False because P and p are different."""
````

````python
print("programming is used to change the world".endswith("world"))
````
