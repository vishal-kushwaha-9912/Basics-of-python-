```python

#Question 4
# Factorial of a number using a loop.


N =3
result =1 
for x in range (1, N+1):
    result = result * x
    print(result)
   
```

```python
#question 6
# Count how many numbers between 1–100 are divisible by 7.
total =0 
for x in range(1,101):
    if x %7 == 0:
        total +=1 
print(total)
```

```python
num = int(input("Enter the number: "))

reverse = 0

for _ in str(num):
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print(reverse)
```
