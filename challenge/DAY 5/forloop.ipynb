```python


# Question 7"
# Reverse a number (e.g., 1234 → 4321) using a loop.


num = int(input("Enter the number: "))

reverse = 0

for _ in str(num):
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print(reverse)
```
