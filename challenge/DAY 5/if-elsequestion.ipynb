```python
#Question 4
# Grade calculator: take marks, print grade (A/B/C/F) using if-elif-else.

user = float(input("Enter your marks"))

if user >=  80:
    print("Grade A")
    
elif user >= 60 :
    print("Grade B")
elif user >= 50:
    print("Grade c")
else:
    print("Grade F")
```

```python
#Question 5
# FizzBuzz: print numbers 1–100, but "Fizz" for multiples of 3, "Buzz" for multiples of 5, "FizzBuzz" for both.

for x in range (1,101):
    if x%3 == 0 and x % 5 == 0:
        print(f"{x} - FizzBuzz")
    elif x %3 == 0:
        print(f"{x} - Buzz")
    elif  x %5 == 0:
        print(f"{x} - Fizz")
    else :
        print(x)
    
```
