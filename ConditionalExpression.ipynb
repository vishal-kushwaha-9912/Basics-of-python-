## Conditional Expression (Ternary Operator )

#### A conditional expression is a short way of making a decision in Python.

Think of it like this:

👉 "If something is true, do this; otherwise, do that."

Real-Life Example

Imagine your mom says:

"If you finish your homework, you can play games; otherwise, you must study."

#### Conditional Expression Format
* variable = value_if_true if condition else value_if_false

#### General Structure

* result = "Yes" if condition else "No"

Python can do the same thing in one line.

##### * Normal if-else statement ::

age = 15

if age >= 13:
    result = "Teenager"
else:
    result = "Child"


##### * Conditional Expression
age = 15

result = "Teenager" if age >= 13 else "Child"

#### Both give the same result:

##### Teenager

#### Why Use It?

✅ Shorter code
✅ Easy for simple decisions
✅ Saves lines of code

Remember

Conditional Expression = One-line if-else

"Yes" if condition else "No"

#### This is also called the Ternary Operator in Python.

```python
# Example 1: Even or Odd
number = 8

result = "Even" if number % 2 == 0 else "Odd"

print(result)
```

```python
# Example 2: Pass or Fail
marks = 75

result = "Pass" if marks >= 40 else "Fail"

print(result)
```

```python
"""Example Covering Different Conditions"""


age = 16
status = "Adult" if age >= 18 else "Minor"

marks = 85
result = "Pass" if marks >= 40 else "Fail"

number = 10
parity = "Even" if number % 2 == 0 else "Odd"

temperature = 35
weather = "Hot" if temperature > 30 else "Cool"

print("Status:", status)
print("Result:", result)
print("Number:", parity)
print("Weather:", weather)
```
