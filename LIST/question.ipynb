📝 Question

Write a Python program that asks the user to enter multiple numbers (including duplicate numbers). Remove all duplicate numbers and print only the unique numbers.

```python

numbers = input("Enter the numbers: ").split()

unique_numbers = []

for num in numbers:
    if num not in unique_numbers:
        unique_numbers.append(num)

print("Unique numbers:", * unique_numbers)
print(f"Numbers enter :{numbers}")
```

How It Works
1) input().split() takes the user's input and creates a list.
2) unique_numbers = [] creates an empty list to store unique values.
3) The for loop checks each number.
4) if num not in unique_numbers: ensures the number hasn't been added already.
5) append() adds only the first occurrence of each number.
6) print(*unique_numbers) prints the numbers separated by spaces.
7) print(f"Numbers enter :{numbers}") :Prints the list of numbers entered by the user.
