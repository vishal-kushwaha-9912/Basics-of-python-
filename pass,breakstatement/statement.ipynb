```python
#>>>>>>>>>> BREAK STATEMENT <<<<<<<
# :A break statement is used to immediately stop a loop and come out of it.



# Real-Life Example

# Imagine you're searching for your friend in a classroom.

# Student 1 ❌
# Student 2 ❌
# Student 3 ❌
# Friend Found ✅
# Stop Searching

# As soon as you find your friend, you stop searching.

# This "stop searching" is like break.



for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
# >>>>Continue Statement <<<<<<<<

# A continue statement is used to skip the current iteration of a loop and move to the next iteration.


# eal-Life Example

# Imagine a teacher is checking attendance:

# Student 1 ✔
# Student 2 ✔
# Student 3 Absent ❌ (Skip)
# Student 4 ✔
# Student 5 ✔

# The teacher skips Student 3 and continues checking the remaining students.

# This is exactly what continue does.

# EXAMPLE :
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
    
    
    # see the output 3 is skipped.
    
    
```

```python
#>>>>>>>>> Pass Statement<<<<<<<<<<<

# The pass statement is a null statement that performs no action and is used as a placeholder for future code.


# Real-Life Example

# Imagine you're writing a book outline:

# Chapter 1 - Written
# Chapter 2 - Written
# Chapter 3 - To be written later

# You leave Chapter 3 blank for now.

# pass is like saying:

# "I'll write this code later."



for i in range(1, 6):
    if i == 3:
        pass
    print(i)
    
    # Notice that pass doesn't change anything—it simply does nothing.
```

```python
# Easy Trick to Remember


# break = Stop 🚪


# continue = Skip ⏭️


# pass = Ignore 🤷‍♂️
```
