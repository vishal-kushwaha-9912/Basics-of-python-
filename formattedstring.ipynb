### Formatted String (f-string) in Python


#### Definition :

#### A Formatted String, also called an f-string, is a special way to insert variables, values, or expressions directly inside a string.


```python
name = "Vishal"

print(f"Hello, {name}")
```

### Why Use f-Strings?

Before f-strings, programmers used:


#### without using of f - string 

```python
#
name = "Vishal"
age = 17

print("My name is " + name + " and I am " + str(age) + " years old.")

#This becomes difficult to read.
```


### Using an f-string:
.

```python
name = "Vishal"
age = 17

print(f"My name is {name} and I am {age} years old.")
```

## Printing Variables

```python
name = "Vishal"
city = "Meerut"

print(f"My name is {name}.")
print(f"I live in {city}.")
```

## Multiple Variables

```python
name = "Vishal"
age = 17

print(f"My name is {name} and I am {age} years old.")
```

### Mathematical Expressions

You can perform calculations inside {}.

```python
a = 10
b = 5

print(f"Sum = {a + b}")
print(f"Product = {a * b}")
```

### Using function 

#### change the output in captil latter 

```python
name = "vishal"

print(f"Uppercase: {name.upper()}")
```
